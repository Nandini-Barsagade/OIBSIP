Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 1: Data Collection


```python
unemployment_data = pd.read_csv('unemployment_data.csv')
```

Step 2: Data Preprocessing

```python
unemployment_data[' Date'] = pd.to_datetime(unemployment_data[' Date'])
```

Step 3: Exploratory Data Analysis

```python
unemployment_rate = unemployment_data[' Estimated Unemployment Rate']
employed = unemployment_data[' Estimated Employed']
labour_participation_rate = unemployment_data[' Estimated Labour Participation Rate']
```

Step 4: Perform data analysis

```python
mean_unemployment_rate = unemployment_rate.mean()
max_employed = employed.max()
min_labour_participation_rate = labour_participation_rate.min()
```

```python
print("Mean Unemployment Rate:", mean_unemployment_rate)
print("Max Employed:", max_employed)
print("Min Labour Participation Rate:", min_labour_participation_rate)
```

Plot the unemployment rate over time

```python
plt.plot(unemployment_data[' Date'], unemployment_rate)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate Over Time')
plt.xticks(rotation=45)
plt.show()
```

Histogram of the unemployment rate distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(unemployment_rate, kde=True)
plt.xlabel("Unemployment Rate")
plt.ylabel("Frequency")
plt.title("Unemployment Rate Distribution")
plt.show()
```

Scatter plot of unemployment rate vs. labor participation rate

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=unemployment_rate, y=labour_participation_rate)
plt.xlabel("Unemployment Rate")
plt.ylabel("Labour Participation Rate")
plt.title("Scatter Plot: Unemployment Rate vs. Labour Participation Rate")
plt.show()
```
